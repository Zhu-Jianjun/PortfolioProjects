```python
import json

aircrafts_data['model'] = aircrafts_data['model'].apply(lambda x: json.loads(x)['en'])
aircrafts_data
```

For the about the above cell: json.loads() function can be used to load JSON data into a Python object. 
This can be useful if one wants to extract specific parts of the JSON data based on the key.
For example, the following code would extract the name and projects properties from the JSON data:

```python
import json

json_data = '[{"name": "Alice", "projects": ["project1", "project2"]}, {"name": "Bob", "projects": ["project3"]}]'

data = json.loads(json_data)
print(type(data)), print(data)
```

```python
for item in data:
    print(item["name"], item["projects"])
```
